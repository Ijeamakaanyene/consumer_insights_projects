```python
import pandas as pd
import numpy as np
```

```python
raw_derm_data = pd.read_csv("Datasets/dermatologistList.csv", low_memory = False)
```

```python
raw_derm_data.head(10)
```

```python
raw_derm_data.columns
```

```python
# Reduce dataframe to columns of interest 
raw_derm_reduced = raw_derm_data.iloc[:, [1, 3, 15, 16, 17, 18, 19, 20]]
```

```python
N = len(raw_derm_reduced.index)
N
```

```python
raw_derm_reduced.isnull().sum() / N * 100
```

```python
# There is about no more than ~2.3% of missing data for most columns. 
# Colums of concern are gender and address, but do not necessarily need gender OR address,
# Street is more important! So will drop those two columns
```

```python
raw_derm_reduced = raw_derm_reduced.drop(["address", "gender"], axis = 1)
```

```python
raw_derm_reduced.head(4)
```

```python
raw_derm_reduced["zipcode_abbrev"] = raw_derm_reduced["zipcode"].str[:5]
```

```python
raw_derm_reduced["zipcode_abbrev"].head(4)
```

```python
derm_deduped = raw_derm_reduced.drop_duplicates(subset = ["fullName", "street", "city", "state", "zipcode_abbrev"])
```

```python
len(derm_deduped.index)
# 10671 people are remaining. But that is a-okay because approx
# ~9.5k dermatologists in the United States
```

```python
derm_deduped[derm_deduped["zipcode_abbrev"].isnull()]
```

```python
# I can't work with people whose zipcodes and city are missing so will remove those!
```

```python
derm_clean = derm_deduped[derm_deduped["zipcode_abbrev"].notnull()]
derm_clean = derm_clean[derm_clean["city"].notnull()]
len(derm_clean.index)
```

```python
derm_clean.head(5)
```

```python
# Identifies people with incorrect cities
derm_clean.loc[derm_clean.city.str.contains("[:0-9:]"), :]
```

```python
# Identifies people with incorrect states
derm_clean.loc[derm_clean.state.str.contains("[:0-9:]"), :]
```

```python
# Identifies people with zipcodes that do not contain numbers
derm_clean.loc[derm_clean.zipcode_abbrev.str.contains("[:0-9:]") == False, :]
```
